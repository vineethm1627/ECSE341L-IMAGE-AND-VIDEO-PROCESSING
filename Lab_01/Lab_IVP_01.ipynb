# Reading and displaying images using :
## 1. PIL
## 2. Matplotlib
## 3. Skimage

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
data_path = "data/rgb_images/"
```

```python
# Reading images using Matplotlib
images_path = []
images_data = []
for file in os.listdir(data_path):
    p = os.path.join(data_path, file)
    images_path.append(p)
    images_data.append(mpimg.imread(p))
```

```python
print(images_path)
print("\n========================================================\n")
print(images_data[0])
```

```python
plt.imshow(images_data[0])
```

```python
for i, img in enumerate(images_data):
    print("Dimension of image {} : {}".format(i + 1, img.shape))
    print("Height of image {} : {}".format(i + 1, img.shape[0]))
    print("Width of image {} : {}".format(i + 1, img.shape[1]))
    print("No. of channels in image {} : {}".format(i + 1, img.shape[2]))
    print("\n========================================================\n")
```

```python
#using PIL
from IPython.display import display
from PIL import Image

img_pil = Image.open(images_path[0])
img_pil = img_pil.resize((250, 250), Image.ANTIALIAS)
#img_pil.show() shows in the desktop app
display(img_pil) #renders the image in the jupyter notebook
print("Height of the image : ", img_pil.size[0])
print("Width of the image : ", img_pil.size[1])
```

```python
#using skimage
from skimage import io
img_sk = io.imread(images_path[0])
io.imshow(img_sk)
io.show()
print("Height of the image : ", img_sk.shape[0])
print("Width of the image : ", img_sk.shape[1])
print("No. of Channels : ", img_sk.shape[2])
```

## Task 2:
Reading and displaying an image in Python without using special libraries such as
OpenCV, PIL or Matplotlib. Write a program that reads a grayscale image in raw format
from a file and save the new image into a new file in raw format. Numpy is the only
standard library that can be used

```python
file_input = open('data/lena_gray.raw')
print(file_input)
```

```python
r, c = 512, 512
img = np.fromfile(file_input, dtype = np.uint8, count = r * c)
print("Size of the image : ", img.size)
```

```python
#converting 1d to 2d
img.shape = (img.size // c, c)
print("2d shape of the image : ", img.shape)
```

```python
#saving the image
img.astype('int8').tofile('saved_image.raw')
file_input.close()
```
