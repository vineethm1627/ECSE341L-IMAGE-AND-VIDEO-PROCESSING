```
#importing the basic libraries

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```
#converting into gray scale image [library]
from PIL import Image

img = Image.open('pic1.jpeg')
img_gray = img.convert('LA')

fig, axs = plt.subplots(nrows = 1, ncols = 2)
axs[0].imshow(img)
axs[1].imshow(img_gray, cmap = 'gray')
```

```
#converting into grayscale image [numpy and matplotlib only]
def rgb2gray(rgb):
    r, g, b = rgb[:, :, 0], rgb[:, :, 1], rgb[:, :, 2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
    return gray


img_g = rgb2gray(np.asarray(img))

fig, axs = plt.subplots(nrows = 1, ncols = 2)
axs[0].imshow(img)
axs[1].imshow(img_g, cmap = 'gray')
```

```
img1 = mpimg.imread('pic1.jpeg')
img2 = mpimg.imread('pic2.jpeg')

print(img1.shape)
print(img2.shape)
print("type of image : ", type(img1))
```

```
img_sub_01 = img1 - img2
img_sub_02 = img2 -img1
img_add = img1 + img2
img_mult = img1 * img2

fig, axs = plt.subplots(2, 2)
axs[0, 0].imshow(img_sub_01)
axs[0, 0].set_title("Subtraction 1")
axs[0, 1].imshow(img_sub_02)
axs[0, 1].set_title("Subtraction 2")
axs[1, 0].imshow(img_add)
axs[1, 0].set_title("Addition")
axs[1, 1].imshow(img_mult)
axs[1, 1].set_title("Multiplication")
```

```
#convert image from one color space to another [RGB to HSV]
from matplotlib.colors import rgb_to_hsv

img_rgb = img
img_hsv = rgb_to_hsv(img_rgb)

fig, axs = plt.subplots(nrows = 1, ncols = 2)
axs[0].imshow(img_rgb)
axs[1].imshow(img_hsv)
```
