```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imsave
from skimage.morphology import erosion, dilation, opening, closing, white_tophat
from skimage.morphology import black_tophat, skeletonize, convex_hull_image
from skimage.morphology import disk, square, diamond
from skimage.color import rgb2gray
from skimage.util import img_as_ubyte
from skimage.morphology import disk
```

```python
img = imread('man_horse.png', as_gray=True)
selem = disk(4)
img_eroded = erosion(img, selem)
img_boundary = np.subtract(img, img_eroded)
```

```python
plt.figure(figsize = (10, 10))

plt.subplot(121)
plt.axis('off')
plt.title("original")
plt.imshow(img, cmap = 'gray')

plt.subplot(122)
plt.axis('off')
plt.title("boundary")
plt.imshow(img_boundary, cmap = 'gray')
```

2. Hole filing using Morphology <br>
Fill the holes in a binary image using morphology. You can use the Python library: <br>
from scipy.ndimage.morphology import binary_fill_holes <br>
and use direct inbuilt function “binary_fill_holes” <br>

```python
from scipy.ndimage.morphology import binary_fill_holes
img = imread('text.png', as_gray = True)
img_hole = binary_fill_holes(img)
```

```python
plt.figure(figsize = (10, 10))

plt.subplot(121)
plt.axis('off')
plt.title("original")
plt.imshow(img, cmap = 'gray')

plt.subplot(122)
plt.axis('off')
plt.title("hole filling")
plt.imshow(img_hole, cmap = 'gray')
```

3. Binary erosion with different size of rectangle (structuring element).

```python
from skimage.morphology import binary_erosion, rectangle
from skimage.morphology import disk

img = imread('man_horse.png', as_gray=True)
img = rgb2gray(img)
print(np.max(img))
# binary thresholding
img[img <= 0.5] = 0
img[img > 0.5] = 1

plt.gray()
plt.figure(figsize=(20,10))

plt.subplot(131)
plt.imshow(img)
plt.title('original', size=20)
plt.axis('off')

plt.subplot(1,3,2)
selem = disk(5)
#Here: img = use the function “binary_erosion” with rectangle size (1,5)
img2 = binary_erosion(img,selem)
plt.imshow(img2)
plt.title('erosion with rectangle size (1,5)', size=20)
plt.axis('off')

plt.subplot(1,3,3)
#Here: img = use the function “binary_erosion” with rectangle size (1,15)
selem = disk(15)
img3 = binary_erosion(img,selem)
plt.imshow(img3)
plt.title('erosion with rectangle size (1,15)', size=20)
plt.axis('off')
plt.show() 
```
